## Part 1: Introduction to Git and GitHub

### What is Git?
**Git** is a version control system â€” think of it as a "Save" button on steroids.

Instead of:
```
my_project_v1.py
my_project_v2.py
my_project_final.py
my_project_final_FINAL.py
my_project_FINAL_final_ACTUAL.py ğŸ˜±
```

Git lets you:
- Save versions with messages: "Add login feature"
- Track who made each change and when
- Go back to any previous version instantly
- Work with teammates without overwriting each other

### What is GitHub?
**GitHub** is the cloud storage for Git repositories.

```
Git (Local)           â†’    GitHub (Cloud)
Your computer              github.com
Tracks versions            Stores all versions online
```

### Git vs GitHub
| Git | GitHub |
|-----|--------|
| Software on your computer | Website hosting your projects |
| Tracks changes locally | Backs up online + enables collaboration |
| Command-line tool | Visual interface (github.com) |
| Works offline | Requires internet to sync |

---

**In human terms:**
- **Git** = Your personal notebook where you write down every change
- **GitHub** = A library where you store copies of all your notebooks for safekeeping

## Part 2: Creating a Repository

### What is a Repository?
A **repository** (or "repo") is a folder that Git tracks. It contains:
- Your code/notebooks
- A hidden `.git/` folder (Git's brain â€” don't touch it!)
- A `.gitignore` file (tells Git what to ignore)
- Typically a `README.md` file (project description)

### Visual:
```
my-project/ (the repository folder)
â”œâ”€â”€ .git/ (Git's hidden folder with version history)
â”œâ”€â”€ .gitignore (tells Git what files to ignore)
â”œâ”€â”€ README.md (project description)
â”œâ”€â”€ notebook.ipynb (your code)
â””â”€â”€ data.csv (data file)
```

### How to Create a Repository (Two Ways)

**Way 1: Create on GitHub.com, then clone to computer** âœ… (what you did!)
1. Go to github.com
2. Click "New" â†’ Create repository
3. Name: `my-project`
4. Check: "Add a README file"
5. Check: "Add .gitignore" â†’ Choose "Python"
6. Click "Create repository"
7. GitHub creates your repo online

**Way 2: Create locally, push to GitHub** (more advanced, skip for now)

### Key Point
After creating a repository on GitHub, you have a folder on GitHub.com.
But it only exists in the cloud â€” you can't edit it directly.
You need to **clone** it to get a copy on your computer.

## Part 3: Cloning a Repository

### What Does "Clone" Mean?
**Clone** = Download a complete copy of a repository from GitHub to your computer.

```
GitHub.com                          Your Computer
my-project (online)     â†’    Clone â†’    my-project/ (local copy)
```

You get:
- All files (notebooks, data, README)
- Complete version history (all commits)
- The `.git/` folder (so Git knows it's a repository)

### Why Clone?
- To work on files locally (in VS Code)
- To have a backup on your computer
- To make changes offline
- To push those changes back to GitHub

### How to Clone

**Step 1: Get the Repository URL**
1. Go to the repository on GitHub.com
2. Click the green "<> Code" button
3. Copy the HTTPS URL (e.g., `https://github.com/lexiloughlin/python-notebooks-practice`)

**Step 2: Clone in Terminal**
```bash
cd ~/Documents
git clone https://github.com/lexiloughlin/python-notebooks-practice
```

**Result:**
```
~/Documents/
â””â”€â”€ python-notebooks-practice/  â† New folder!
    â”œâ”€â”€ .git/
    â”œâ”€â”€ .gitignore
    â”œâ”€â”€ README.md
    â””â”€â”€ my_first_analysis.ipynb  â† Your files
```

### After Cloning
- Open folder in VS Code
- You can now edit files locally
- You can commit changes
- You can push changes back to GitHub

### Key Point
**Clone creates a connection between your computer and GitHub.**
Git remembers:
- Where you cloned from (the "remote")
- How to send changes back up to GitHub (push)
- How to download changes from GitHub (pull)

## Part 4: Making a Commit

### What is a Commit?
A **commit** is a snapshot of your project at a specific point in time.

Think of it like:
- **Save in a video game** = One checkpoint you can return to
- **Commit** = One checkpoint in your project history

### Visual Timeline of Commits
```
Commit 1          Commit 2          Commit 3
"Initial files"   "Add function"    "Fix bug"
    â†“                 â†“                 â†“
Time: 9:00 AM    Time: 9:30 AM    Time: 10:00 AM
```

Each commit has:
1. **Message**: What you changed (e.g., "Add login feature")
2. **Author**: Who made the change (you)
3. **Timestamp**: When it happened
4. **Hash**: A unique ID (e.g., `a3f8b2c`)

### The 3-Step Commit Process

**Step 1: Modify Your Files**
```
You edit: my_analysis.ipynb
You save it (Ctrl+S)
```

**Step 2: Stage Changes**
Staging = "Select which files to include in this commit"

In VS Code:
1. Open Source Control (Ctrl+Shift+G)
2. See files under "Changes"
3. Click "+" next to file to stage it (moves to "Staged Changes")

```bash
# Or in terminal:
git add my_analysis.ipynb
```

**Step 3: Commit with Message**

In VS Code:
1. Type message in the box: "Add data analysis section"
2. Click checkmark to commit

```bash
# Or in terminal:
git commit -m "Add data analysis section"
```

### Good Commit Messages
âœ… **Good:**
- "Add login feature"
- "Fix bug in calculator function"
- "Update documentation"
- "Add data visualization"

âŒ **Bad:**
- "update"
- "asdf"
- "fix"
- "work in progress"

### Key Point
**Commits are local only â€” they live on YOUR computer in the `.git/` folder.**
To share them with GitHub, you need to **push** (next step!).

## Part 5: Pushing Changes to GitHub

### What is Push?
**Push** = Upload your local commits to GitHub.

```
Your Computer (Local)     â†’    Push â†’    GitHub.com (Remote)
Commits in .git/               Upload       Commits visible online
```

After pushing:
- Your changes are safe in the cloud
- Your teammates can see your work
- You have a backup

### Before You Can Push
You must have:
1. âœ… Created a repository on GitHub
2. âœ… Cloned it to your computer
3. âœ… Made commits locally
4. âœ… Configured Git with your name and email (one-time setup)

### How to Push

**In VS Code (Easiest):**
1. Open Source Control (Ctrl+Shift+G)
2. Click the "â‹¯" menu (three dots)
3. Select "Push"
4. Wait for success message

**Or use the Sync button:**
- Bottom of VS Code, you'll see circular arrows (â†•ï¸)
- Click to sync (push + pull at once)

**In Terminal:**
```bash
git push
```

### What Happens After Push
1. Git uploads your commits to GitHub
2. GitHub updates your repository online
3. You can see your changes on github.com
4. Your teammates can see your work

### Verify Your Push
1. Go to github.com
2. Navigate to your repository
3. Your files appear in the file list
4. Click "Commits" to see your commit history

### Key Point
**If it's not on GitHub, it didn't happen.**
Always verify your push was successful by checking github.com!

## Part 6: Pulling Changes from GitHub

### What is Pull?
**Pull** = Download changes from GitHub to your computer.

```
GitHub.com (Remote)     â†’    Pull â†’    Your Computer (Local)
New commits from teammate      Download   Updates your files
```

### Why Pull?
When working with teammates:
1. Teammate makes changes â†’ commits â†’ pushes to GitHub
2. You need those changes on your computer
3. You pull to download their work
4. Now you both have the latest version

### The Golden Rule: ALWAYS PULL BEFORE YOU EDIT

```
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘  PULL â†’ EDIT â†’ TEST â†’ COMMIT â†’ PUSH  â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
```

**Why?** If you skip the pull:
1. Teammate pushed changes to GitHub
2. You have old version on your computer
3. You make different changes
4. You try to push â†’ Git says "Your version is old!"
5. You have to fix the conflict

### How to Pull

**In VS Code (Easiest):**
1. Open Source Control (Ctrl+Shift+G)
2. Click "â‹¯" menu
3. Select "Pull"
4. Your files update

**Or click Sync button (â†•ï¸):**
- Syncs your local changes up and new changes down

**In Terminal:**
```bash
git pull
```

### What Happens When You Pull
1. Git checks what's new on GitHub
2. Downloads any new commits
3. Merges them into your local version
4. Your files are now up to date

### Example Workflow (With a Teammate)

**Tuesday morning:**
```
You: git pull  (get latest version)
Teammate: git pull  (get same latest version)
```

**Tuesday 10 AM:**
```
You: Add section 1 â†’ commit â†’ push
GitHub now has your section 1
```

**Tuesday 10:15 AM:**
```
Teammate: git pull (gets your section 1)
Teammate: Add section 2 â†’ commit â†’ push
```

**Tuesday 10:30 AM:**
```
You: git pull (gets teammate's section 2)
Both sections now on your computer!
```

### Key Point
**Git + GitHub = Synchronized teamwork**
- Push your work to GitHub â†’ teammates can pull it
- Teammates push their work â†’ you pull it
- Everyone stays on the same page!
