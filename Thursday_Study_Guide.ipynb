## Part 1: What is Forking?

### The Big Idea
**Fork** = Make YOUR OWN copy of someone else's repository on GitHub

```
Instructor's Repo (GitHub)          Your Fork (GitHub)              Your Computer
github.com/prof/assignment    â†’    github.com/YOU/assignment    â†’    you edit here
      (read-only)                       (you own this!)               (and push back)
```

### Why Fork?
- Instructor posts assignment template â†’ YOU fork it â†’ YOU modify YOUR copy
- Instructor can't push to your repo â†’ YOU have full control
- Professional way to contribute to open-source projects

### When to Fork vs Clone

| Your Situation | What to Do |
|---|---|
| It's YOUR repo | Clone directly (no fork needed) |
| You're a collaborator | Clone directly (instructor added you) |
| Submitting homework from template | **Fork first, then clone your fork** |
| Contributing to someone's open-source | **Fork first, then clone your fork** |

## Part 7: Merge Conflicts (Don't Panic!)

### What is a Merge Conflict?
You and your partner **edit the same line** in the same file â†’ Git says "I don't know which version to keep!"

### Example
```python
<<<<<<< HEAD (your version)
def add(a, b):
    return a + b + 10

=======
def add(a, b):
    return a + b

>>>>>>> origin/main (partner's version)
```

### How to Fix It in VS Code
VS Code shows 4 buttons above the conflict:

1. **"Accept Current Change"** â†’ Keep YOUR version
2. **"Accept Incoming Change"** â†’ Keep PARTNER's version
3. **"Accept Both Changes"** â†’ Keep both (careful!)
4. **Edit manually** â†’ Write a new version that combines both

**After choosing:**
1. Save the file
2. Commit: `"Resolve merge conflict"`
3. Push

### Best Way to Avoid Conflicts
âœ… Work on **different cells/sections** when possible
âœ… **Pull before you edit**
âœ… **Clear outputs before committing** (reduces JSON conflicts in notebooks)
âœ… **Communicate:** "I'm working on the add function"
